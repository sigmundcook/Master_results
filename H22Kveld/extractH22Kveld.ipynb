```python
import pandas as pd

df = pd.read_csv("H22_kveld.csv", sep = ";" , encoding = "utf-8")
df_keep = ["ext_inspera_questions_5_ext_inspera_candidateResponses_0_ext_inspera_response", "ext_inspera_questions_6_ext_inspera_candidateResponses_0_ext_inspera_response", "ext_inspera_questions_17_ext_inspera_candidateResponses_0_ext_inspera_response"]
df = df[df_keep]

df.head()
print(df.columns)
```

```python
for y in range (len(df.columns)):
    # column = df.columns[y]
    for i in range(len(df.index)):
        filename= "H22Kveld/oppgave1/export" + str(i) + ".py"
        try:
            code_str = str(df.at[i, 'ext_inspera_questions_5_ext_inspera_candidateResponses_0_ext_inspera_response'])
            # new_text = code_str.split("# Skriv koden din mellom her")
            # new_text2 = new_text[1].split("# IKKE endre koden")
            code_str = code_str.split('def')
            print(code_str[1]) 
            with open(filename, "w", encoding="utf-8") as file:
                file.write('import numpy as np \ndef' + code_str[1] + "\n")
        except:
            print("Error splitting string in " + str(i))
```
